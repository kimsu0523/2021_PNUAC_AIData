### **리스트(list)를 시리즈(Series) 변경**

#### pd.Series()

````python
lt = [4.1, 5.5, 9.0, 15.0, 19.0 ,21.3]

#리스트를 시리즈로 변경하기
slt = pd.Series(lt)
print(slt)
print(type(slt))
````

#### 리스트를 변경한 series index, values

````python
#시리즈 인덱스, 키, 값 출력
print('index :', slt.index)
print('index :', slt.keys())
print('value :', slt.values)
````

### **시리즈(Series) 인덱싱(indexing)**
* 인덱스를 가지고 시리즈에서 특정 항목 추출



````python
ltk = ['1월', '2월', '3월', '4월', '5월', '6월']
lt = [4.1, 5.5, 9.0, 15.0, 19.0 ,21.3]
slt = pd.Series(lt)
slt2 = pd.Series(lt, index=ltk)
print(slt)
print(slt2)
````

````python
#slt 시리즈에서 인덱스가 2인 항목 값 가져오기
slt[2]
````

````python
#slt2 시리즈에서 3월의 값가져오기
slt2['3월']
````

### **시리즈(Series) 슬라이싱(Slicing)**
* 시리즈 일부를 추출하여 시리즈로 반환
* 콜론(;), 쉼표(,) 사용
* 조건 사용

#### 콜론(:), 쉼표(,) 사용

````python
#인덱스가 1에서 2인 자료 추출
sub1 = slt[1:3]
print(sub1)
````

````python
#인덱스가 1월에서 3월인 자료 추출
sub2 = slt2['1월':'3월']
print(sub2)
````

````python
#인덱스가 1,3,5인 자료 추출
sub1_1 = slt[[1, 3, 5]]
print(sub1_1)
````

````python
##인덱스가 2월,4월, 6월인 자료 추출
sub2_1 = slt2[['2월', '4월', '6월']]
print(sub2_1)
````

### **시리즈(Series) 연산**
* 시리즈(Series) 길이가 같고 인덱스가 같은 것끼리 연산
 + 더하기(+) : 각 값 더하기 
 + 곱하기(*) : 각 값 곱하기

````python
print(slt)
print(slt2)
````

````python
#인덱스가 다른 경우
print(slt + slt2)
````

````python
slt.index = slt2.index
print(slt)
print(slt2)
````

````python
#인덱스가 같은 경우
print(slt + slt2)
print(slt * slt2)
````

### **자주 사용되는 메소드**
+ equals() : 시리즈와 시리즈가 같은지 비교, 키와 값이 모두 같아야 함
+ replace() : 특정값을 가진 요소 값을 변경
+ isin() : 시리즈에 포함된 값이 있는지 확인
+ drop_duplicates() : 중복값이 없는 시리즈 반환
+ append() : 2개이상의 시리즈 연결
+ sort_index(ascending=True) : 인덱스로 정렬
+ sort_values(ascending=False) : 값으로 정렬
+ to_frame() : 시리즈를 데이터프레임으로 변경

````python
dt = {'1월':4.1, '2월':5.5, '3월':9.0, '4월':15.0, '5월':19.0 ,'6월':21.3}
sdt1 = pd.Series(dt)  
sdt2 = pd.Series(dt.values()) 
````

````python
print('[시리즈1]')
print(sdt1)
print('[시리즈2]')
print(sdt2) 
````

````python
print('비교 :', sdt1.equals(sdt2))
````

````python
print('변경 :')

#5.5 값을 15로 변경
sdt1 = sdt1.replace(5.5, 15)
print(sdt1)
````

````python
sdt1 = sdt1.replace(15, 15.5)
print(sdt1)
````

````python
print('포함:')
#값을 15.5, 9.0을 가지는 자료 추출
print(sdt1[sdt1.isin([15.5, 9.0])])
````

````python
print('중복제거:')
print(sdt1.drop_duplicates()) 
print('중복제거반영확인:')
#중복제거하여 시리즈에 반영
sdt1 = sdt1.drop_duplicates()
print(sdt1)
````

````python
#시리즈 연결하기
print('연결:')
print(sdt1.append(sdt2))
````

````python
print('정렬(값):')
#시리즈 값으로 정렬하기
print(sdt1.sort_values(ascending=False))
print('정렬(인덱스):')
#시리즈 인덱스로 역순으로 정렬하기
print(sdt1.sort_index(ascending=False))
````

````python
print('데이터프레임변환:')
#시리즈 데이터 프레임으로 변환하기
sdt1.to_frame()
````

1. 딕션너리 시리즈로 변환
2. 학과를 입력 받아서 해당학과의 후보순위, 단 입력종료 메시지를 추가하여 종료가 입력되지 않으면 계속 입력
   * 결과예시)
   ``` 
    학과입력 =>
일어학과
해당 학과는 존재하지 않습니다.
종료하시겠습니까?(종료:y)n
학과입력 =>
일어일문학과
일어일문학과 추가 합격 순위 : 후보 8
종료하시겠습니까?(종료:y)y
  ```
3. 후보순위가 3번 이하인 학과와 후보순위가 20번 이상인 학과 조회
   * 조건식을 이용하는 방법
   * 메소드를 이용하는 방법 
4. 평균 추가 합격 순위
   * 출력예) 평균 추가 합격 순위 : 7위
5. 학과 명칭대로 정렬하여 시리즈에 반영
 

````python
dt = {'국어국문학과':7,	'중어중문학과':4,	'일어일문학과':8,	'영어영문학과':30,	'불어불문학과':5,	'독어독문학과':4,	'노어노문학과':5,	'한문학과':6,	'언어정보학과':3,	'사학과':7,	'철학과':5,	'고고학과':9}
````

````python
#학과 후보순위 2트
dts = pd.Series(dt)
lt = dts.index.to_list()
s = 1
while(s == 1) :
  i = 0
  print('학과입력 =>')
  hak = input()
  if hak in lt :
    print(hak, ' 추가 합격 순위 : 후보 ', dt[hak])
    stop = input('종료하시겠습니까?(종료:y)')
    if stop == 'y' or stop == 'Y':
      print('프로그램을 종료합니다.')
      break
    else:
      continue
  else:
    print('해당 학과는 존재하지 않습니다.')
    stop = input('종료하시겠습니까?(종료:y)')
    if stop == 'y' or stop == 'Y':
      print('프로그램을 종료합니다.')
      break
    else:
      continue
````
